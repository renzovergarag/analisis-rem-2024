### Limpieza y Agrupación de Datos por Centro de Salud

```python
def clean_data(df):
    df = df.groupby(['GENERO']).agg(CENTRO_count=('NOMBRE_CENTRO', 'count')).reset_index()
    # Ordenar por columna: 'CENTRO_count' (descendente)
    df = df.sort_values(['CENTRO_count'], ascending=[False])
    return df
```
